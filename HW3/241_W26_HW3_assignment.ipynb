### **Problem 1: Basis Vectors**
Consider the following set of vectors:

$\begin{pmatrix}2\\0\\1\end{pmatrix},\;\begin{pmatrix}0\\3\\0\end{pmatrix},\;\begin{pmatrix}-2\\0\\4\end{pmatrix}$

a) Decompose the vector $\begin{pmatrix}4 \\ 3 \\ -1\end{pmatrix}$ onto these vectors by setting up and solving a linear system to find the coefficients. You can use Python's and NumPy's builtin functions to solve the linear system. Print the resulting vector.

```python
import numpy as np
#Code Here
v1 = np.array([2,0,1], dtype=float)
v2 = np.array([0,3,0], dtype=float)
v3 = np.array([-2,0,4], dtype=float)
v = np.array([4,3,-1], dtype=float)

V = np.column_stack([v1,v2,v3])
coeffs = np.linalg.solve(V,v)

print(coeffs)
```
